```python
# Read CSV as pandas data frame
import pandas as pd
df1=pd.read_csv('/content/drive/MyDrive/CompAndViz/average_price.csv')
df2=pd.read_csv('/content/drive/MyDrive/CompAndViz/median_price.csv')
df3=pd.read_csv('/content/drive/MyDrive/CompAndViz/new_houses_sold.csv')
df4=pd.read_csv('/content/drive/MyDrive/CompAndViz/mortgage.csv')
```

```python
import pandas as pd
df6=pd.read_csv('/content/drive/MyDrive/CompAndViz/1bhk.csv')
df7=pd.read_csv('/content/drive/MyDrive/CompAndViz/2bhk.csv')
df8=pd.read_csv('/content/drive/MyDrive/CompAndViz/3bhk.csv')
```

```python
df6
```

```python
df6 = df6.drop(columns=['RegionID','SizeRank','RegionType','StateName'])
df7 = df7.drop(columns=['RegionID','SizeRank','RegionType','StateName'])
df8 = df8.drop(columns=['RegionID','SizeRank','RegionType','StateName'])
```

```python
# Convert from wide to long format
df_1bhk = pd.melt(df6, id_vars=['RegionName'], var_name='Date', value_name='1bhk_home_value')

print("\nLong DataFrame:")
print(df_1bhk)
```

```python
df_1bhk
```

```python
df_2bhk = pd.melt(df7, id_vars=['RegionName'], var_name='Date', value_name='2bhk_home_value')

print("\nLong DataFrame:")
print(df_2bhk)
```

```python
df_3bhk = pd.melt(df8, id_vars=['RegionName'], var_name='Date', value_name='3bhk_home_value')

print("\nLong DataFrame:")
print(df_3bhk)
```

```python
df1
```

```python
df1.rename(columns={'ASPUS': 'avg_price'}, inplace=True)
df2.rename(columns={'MSPUS': 'median_price'}, inplace=True)
df3.rename(columns={'NHSUSSPTQ': 'new_houses_sold'}, inplace=True)
df4.rename(columns={'BOGZ1FL073165103Q': 'mortgage'}, inplace=True)
```

```python
df1
```

```python
# Merge the first two DataFrames
df_merged = pd.merge(df_1bhk, df_2bhk, on=['RegionName', 'Date'], how='inner')

# Now merge the result with the third DataFrame
df_houses = pd.merge(df_merged, df_3bhk, on=['RegionName', 'Date'], how='inner')

# Print the final merged DataFrame
print("\nFinal Merged DataFrame:")
print(df_houses)
```

```python
df_houses
```

```python
df_houses.to_csv('HomeValue.csv')
```

```python
merged_df = df1.merge(df2, on='DATE') \
                .merge(df4, on='DATE')

print(merged_df)
```
